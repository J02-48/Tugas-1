```python
radius = float(input("Masukkan jari-jari lingkaran (tidak usah pakai cm): "))
```

```python
area = 22/7 * radius * radius
```

```python
jwbn = "Luas lingkaran dengan jari-jari {} cm adalah {:.2f} cm\u00b2.".format(radius, area)
```
